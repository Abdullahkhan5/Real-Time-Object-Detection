## Object Detection Network

#### I have used the Mobile-Net-SSD Network

```python
import cv2
```

## Capturing the Video/Image

```python
# img = cv2.imread('lena.png')
capture = cv2.VideoCapture(0)
capture.set(3,640)
capture.set(4,480)
```

## Setting the Confidence Threshold

```python
Threshold = 0.6
```

## Importing names of objects from Dataset

#### I have used the Coco Dataset which is trained on about 91 different objects

```python
objectClassesNames= []
objectsClassesFile = 'coco.names'
with open(objectsClassesFile,'rt') as f:
    objectClassesNames = f.read().rstrip('\n').split('\n')
```

```python
print(objectClassesNames)
```

## Importing Network Configuration file and Trained Weights file

```python
configurationFilePath = './Object_Detection_Files/ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
weightsFilePath = './Object_Detection_Files/frozen_inference_graph.pb'
```

## Processing for the Network

```python
network = cv2.dnn_DetectionModel(weightsFilePath,configurationFilePath)
```

```python
##Suggested Configurations
network.setInputSize(320,320)
network.setInputScale(1.0/127.5)
network.setInputMean((127.5,127.5,127.5))
network.setInputSwapRB(True)
```

## Detecting the Objects

```python
while True:
    success,image = capture.read()
    objectsClassIds , confidence , boundingBox = network.detect(image,confThreshold = Threshold)
    print(objectsClassIds,boundingBox)
    
    if len(objectsClassIds) != 0:
        for objsId, confds, box in zip(objectsClassIds.flatten(),confidence.flatten(),boundingBox): ##For referring the three lists created above
            cv2.rectangle(image,box,color=(0,255,0),thickness=3)
            cv2.putText(image,objectClassesNames[objsId-1].upper(),(box[0],box[1]-10),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),1)
            cv2.putText(image,str(round(confds*100,2)),(box[0],box[1]+30),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),1)
                             
    cv2.imshow('Output Picture',image)
    cv2.waitKey(1)
```
